```python
import random
import math
import matplotlib.pyplot as plt
for i in range(3, 7):
    n = 10**i
    count = 0
    for _ in range(n):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        dist = math.sqrt(x**2 + y**2)
        if dist < 1:
            count += 1
            if n == 10**4:
                plt.scatter(x, y, color='blue', s = 0.75)
        elif n == 10**4:
            plt.scatter(x, y, color='red', s = 0.75)
    pi = 4*count/n
    print(f"n = 10 ^ {i} pi = {pi:.4f} error = {abs(100*(math.pi - pi)/math.pi):.4f}%")
```
